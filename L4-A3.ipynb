```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic Dataset.csv")
df.tail()
```

```python
df.dtypes
```

```python
num_label = ["PassengerId", "Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"]
```

```python
# Dropping all categorical columns
num_data = df.drop(["Name", "Ticket", "Cabin", "Embarked", "Gender"], axis=1)
```

```python
# Boxplot of all numerical data
for label in num_label:
    plt.boxplot(num_data[label])
    print("Distribution of:", label)
    plt.show()
```

```python
# Scaling the dataset to standardize the data

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
num_data_scaled = scaler.fit_transform(num_data)
print(num_data_scaled)
```
