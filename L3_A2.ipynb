```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic Dataset.csv")
df.head(10)
```

```python
# Boxplot of column Age
plt.boxplot(df["Age"], vert=False)
plt.title('Age Box Plot')
plt.xlabel("Age")
plt.show()
```

```python
# Boxplot of column Fare
plt.boxplot(df["Fare"], vert=False)
plt.title('Fare Box Plot')
plt.xlabel("Fare")
plt.show()
```

```python
# Boxplot of column Pclass
plt.boxplot(df["Pclass"], vert=False)
plt.title('Pclass Box Plot')
plt.xlabel("Pclass")
plt.show()
```
