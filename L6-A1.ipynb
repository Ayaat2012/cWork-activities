```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic Dataset.csv")
df.head()
```

```python
# Any null data?
df.isnull().sum()
```

```python
df.info()
```

```python
# Passengers belonging from which gender survived the most?
sns.countplot(x=df["Gender"], hue=df["Survived"], palette='winter')
plt.show()
```

```python
# Passenger belonging to which Pclass survived the most?
sns.countplot(x=df["Pclass"], hue=df["Survived"], palette='spring')
plt.show()
```

```python
# Highest num of passengers belonging to which Age?
sns.displot(x=df["Age"], kde=True, bins=40)
```

```python
# Highest num of passengers belonging to which gender?
sns.countplot(x=df["Gender"])
```

```python
# Is SibSp corr/asso with Survived feature?
sns.countplot(x=df["Survived"], hue=df["SibSp"], palette="summer")
```

```python
# Is Parch corr/asso with Survived feature?
sns.countplot(x=df["Survived"], hue=df["Parch"], palette="mako")
```

```python
# Is Fare having normal distribution?
sns.displot(x=df["Fare"], kde=True)
```

```python
# Checking the age group of majority of ppl belonging to Pclass1
sns.boxplot(x=df["Pclass"], y=df["Age"], color='pink')
```

```python
# Finding the corr of all the features with target variable='Survived'
sns.heatmap(df.corr(numeric_only=True), annot=True)
```
