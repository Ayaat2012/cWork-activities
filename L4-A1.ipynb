```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic Dataset.csv")
df.head()
```

```python
df.dtypes
```

```python
# Nominal category and Ordinal category

nom_data = ["Name", "Ticket", "Cabin"]
ord_data = ["Embarked", "Gender"]
```

```python
# Gender
df["Gender"].value_counts()
```

```python
# Label Encoding for column Gender

gen_cat = ["male", "female"]
df["Gender"] = pd.Categorical(df["Gender"], gen_cat, ordered=True)
median_index = np.median(df["Gender"].cat.codes)
median_gen = gen_cat[int(median_index)]
print("Median of column Gender:", median_gen)
```

```python
# Embarked
df["Embarked"].value_counts()
```

```python
# Label Encoding for column Embarked

em_cat = ["S", "C", "Q"]
df["Embarked"] = pd.Categorical(df["Embarked"], em_cat, ordered=True)
median_index = np.median(df["Embarked"].cat.codes)
median_em = em_cat[int(median_index)]
print("Median of column Embarked:", median_em)
```
