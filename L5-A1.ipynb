```python
import pandas as pd
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic Dataset.csv")
df.head()
```

```python
# Finding the minimum age
min_age = df["Age"].min()
print("Minimum Age:", min_age)
```

```python
# Finding the maximum Age
max_age = df["Age"].max()
print("Maximum  Age:", max_age)
```

```python
# Binning the data
bins = [0, 15, 30, 45, 60, 75]
df['binned_age'] = pd.cut(df["Age"], bins)
print(df[['binned_age', 'Age']].head())
```

```python
# Labels
age_labels = ['Young', 'Young-Adult', 'Middle Aged', 'Middle-Older Aged', 'Senior']

df['binned_age'] = pd.cut(df["Age"], bins, labels=age_labels)
df['binned_age'].value_counts().plot(kind='bar')

# Bar plot
plt.title("Dance Class Age Distribution")
plt.xlabel("Ages")
plt.ylabel("Count")
plt.xticks(rotation=30, fontsize=14)
plt.show()
```

```python
# More distribution plots
labels = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
for label in labels:
    print("Distribution of:", labels)
    sns.displot(df[label], kde=True)
    plt.show()
    print('Skewness :', df[label].skew())
```

```python
# Reducing skewness
df['log_SibSp'] = np.log1p(df['SibSp'])
print(df['log_SibSp'])
df['log_Parch'] = np.log1p(df['Parch'])
print(df['log_Parch'])
df['log_Fare'] = np.log1p(df['Fare'])
print(df['log_Fare'])
```
