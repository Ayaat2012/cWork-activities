```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic Dataset.csv")
df.head()
```

```python
# Setting the bg
sns.set_style("whitegrid")
```

```python
# Countplot
sns.countplot(x=df["Survived"], color='pink')
```

```python
# More countplots
sns.countplot(x=df["Gender"], hue=df["Survived"], palette='spring')
```

```python
# Reverse of the previous countplot
sns.countplot(x=df["Survived"], hue=df["Gender"], palette="summer")
```

```python
# Extra customization
sns.countplot(x=df["Embarked"], palette="winter")
plt.xticks(fontsize=17, rotation=60)
```
