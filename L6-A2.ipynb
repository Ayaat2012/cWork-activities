```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris Dataset.csv")
df.head()
```

```python
df.info()
```

```python
# Statistical info
df.describe()
```

```python
# Create Boxplot of all features and checking for outliers
labels = ["Id", "SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]

# Boxplot
for label in labels:
    print("Distribution of:", label)
    plt.boxplot(df[labels])
    plt.show()
```

```python
# Checking the corr between all features
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

```python
# Checking if any features are skewed or not
labels = ["Id", "SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]

for label in labels:
    print("Distribution of:", label, "using displot.")
    print("Skewness of:", df[label].skew())
    sns.displot(df[label], kde=True)
    plt.show()
```

```python
plt.scatter(x=df["SepalLengthCm"], y=df["PetalLengthCm"], color='deeppink')
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.title("Distribution of SepalLengthCm and PetalLengthCm")
plt.show()
```

```python
plt.scatter(x=df["SepalWidthCm"], y=df["PetalWidthCm"], color='pink')
plt.xlabel("SepalWidthCm")
plt.ylabel("PetalWidthCm")
plt.title("Distribution of SepalWidthCm and PetalWidthCm")
plt.show()
```
