```python
####**Import Libraries**

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import dataset
df = pd.read_csv('IMDB Dataset.csv')
df.head()
```

```python
"""####**Check Null Values**"""
df.isnull().sum()
```

```python
"""**No null values present -**
#### **Plot Histogram for Runtime**
"""

plt.hist(df['Runtime'], color='skyblue')
plt.xlabel("Runtime")
plt.ylabel("Count of movies")
```

```python
"""#### **Plot Histogram for IMDB_Rating**"""

plt.hist(df['IMDB_Rating'], color='pink')
plt.xlabel("IMDB Rating")
plt.ylabel("Count of movies")
```

```python
"""#### **Define parameter bins_runtime for Runtime and plot histogram for it"""

df['Runtime'].unique()

bins_time = np.arange(80, 230, 10)
plt.hist(df['Runtime'], edgecolor="black", bins=bins_time, color='indigo')
plt.xlabel("Rntime")
plt.ylabel("Count of movies")
```

```python
"""#### **Define parameter bins_rating for Runtime and plot histogram for it"""

df['IMDB_Rating'].unique()

bins_time = np.arange(8, 10, 0.20)
plt.hist(df['IMDB_Rating'], edgecolor="black", bins=bins_time, color='purple')
plt.xlabel("IMDB Rating")
plt.ylabel("Count of movies")
```
