```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Titanic Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Quantile of  column Age
q1_age = np.quantile(df['Age'], 0.25)
q2_age = np.quantile(df['Age'], 0.50)
q3_age = np.quantile(df['Age'], 0.75)

print("Q1 of column Age:", q1_age)
print("Q2 of column Age:", q2_age)
print("Q3 of column Age:", q3_age)
```

```python
# IQR of column Age
iqr_age = q3_age - q1_age
print("IQR of column Age:", iqr_age)
```

```python
# Quantile of column Fare
q1_fare = np.quantile(df['Fare'], 0.25)
q2_fare = np.quantile(df['Fare'], 0.50)
q3_fare = np.quantile(df['Fare'], 0.75)

print("Q1 of column Fare:", q1_fare)
print("Q2 of column Fare:", q2_fare)
print("Q3 of column Fare:", q3_fare)
```

```python
# IQR of column Fare
iqr_fare = q3_fare - q1_fare
print("IQR of column Fare:", iqr_fare)
```

```python
# Histogram of column Age
plt.hist(df['Age'], bins=14, color='pink')
plt.title("Passenger vs Age")
plt.xlabel("Age")
plt.ylabel("Passenger")
plt.show()
```

```python
# Histogram of column Fare

bins = np.arange(0, 250, 20)

plt.hist(df['Fare'], bins=bins, color='magenta')
plt.title("Passenger vs Fare")
plt.xlabel("Fare")
plt.ylabel("Passenger")
plt.xticks(bins)
plt.show()
```
