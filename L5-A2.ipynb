```python
import pandas as pd
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic Dataset.csv")
df.head()
```

```python
# A boxplot
sns.boxplot(data = df, x = 'Embarked', y = 'Age', color = 'pink')
plt.show()
```

```python
# Relation between two quantitative data
plt.scatter(x=df['Fare'], y=df['Survived'], color='magenta')
plt.xlabel("Fare")
plt.ylabel("Survived")
plt.show()
```

```python
# Scatter plot for Parch and Survived
plt.scatter(x=df['Parch'], y=df['Survived'], color='purple')
plt.xlabel("Parch")
plt.ylabel("Survived")
plt.show()
```

```python
# Another scatter plot
plt.scatter(x=df['SibSp'], y=df['Survived'], color='indigo')
plt.xlabel("SibSp")
plt.ylabel("Survived")
plt.show()
```

```python
# Association between categorical data
asso_cat = pd.crosstab(df['Gender'], df["Embarked"])
print(asso_cat)
```
